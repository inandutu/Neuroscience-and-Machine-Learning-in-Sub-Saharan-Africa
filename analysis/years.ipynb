```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
import os
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
root_dir = "/content/gdrive/My Drive/AI-neuroscience-knowledge-discovery/Projects/Project: Neuroscience and AI in Africa/Intersection of NS & CS/Synthesis/analysis/"
os.listdir(root_dir)
```

```python
years_published = pd.read_excel(root_dir+'years.xlsx')
years_published
```

```python
years = years_published['Year']
years
```

```python
# Count number of publications per year
year_counts = Counter(years)

# Sort years
sorted_years = sorted(year_counts.keys())
publication_counts = [year_counts[year] for year in sorted_years]

# Plotting
fig, ax = plt.subplots(figsize=(6.93, 4))  # width = 17.6 cm
ax.plot(sorted_years, publication_counts, marker='o', linestyle='-', color='teal')

# Add data labels
for x, y in zip(sorted_years, publication_counts):
    ax.text(x, y + 0.2, str(y), ha='center', fontsize=9)

# Labeling
ax.set_xlabel("Year")
ax.set_ylabel("Number of Publications")
ax.set_title("Publications per Year")
ax.grid(True)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

# Save and display the figure
plt.savefig(root_dir+"years.png", dpi=1200, bbox_inches="tight")
plt.show()
```
