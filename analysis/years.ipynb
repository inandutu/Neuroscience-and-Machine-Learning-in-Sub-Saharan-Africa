```python
years_published = pd.read_excel(root_dir+'years.xlsx')
years_published
```

```python
years = years_published['Year']
years
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Count number of publications per year
year_counts = Counter(years)

# Sort years
sorted_years = sorted(year_counts.keys())
publication_counts = [year_counts[year] for year in sorted_years]

# 👉 Convert to cumulative counts
cumulative_counts = np.cumsum(publication_counts)

# Plotting
fig, ax = plt.subplots(figsize=(7.5, 8.75)) # width_inches = 7.5   # 19.05 cm, and height_inches = 8.75 # 22.23 cm
ax.plot(sorted_years, cumulative_counts, marker='o', linestyle='-')

# Add data labels
for x, y in zip(sorted_years, cumulative_counts):
    ax.text(x, y + 0.2, str(y), ha='center', fontsize=9)

# Labeling
ax.set_xlabel("Year")
ax.set_ylabel(" Number of Publications")
ax.set_title(" Publications Over Time")
ax.grid(True)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

# Save and display the figure
plt.savefig(root_dir+"years.tiff", dpi=600, bbox_inches="tight", pil_kwargs={"compression": "tiff_lzw"})
plt.show()
```
