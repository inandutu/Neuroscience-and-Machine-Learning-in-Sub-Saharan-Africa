```python
df = pd.read_excel(root_dir+'instituions.xlsx')
df
```

```python
institutions = df['Institutions']
```

```python
# Step 2: Count the frequency of each institution
counts = df['Institutions'].value_counts()
```

```python
counts = Counter(institutions)

labels = list(counts.keys())
sizes = list(counts.values())

# Choose a colormap with enough distinct colors (tab20 has 20 distinct colors)
cmap = plt.get_cmap('tab20')
num_colors = len(labels)

# Generate colors from the colormap
colors = cmap(np.linspace(0, 1, num_colors))

fig, ax = plt.subplots(figsize=(8, 6))

wedges, texts, autotexts = ax.pie(
    sizes,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops={'fontsize': 10, 'color': 'white'}
)

ax.legend(
    wedges,
    labels,
    title="Institutions",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=10
)

ax.axis('equal')

# Save and display the figure
plt.savefig(root_dir+"institutions.TIFF", dpi=1200, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import numpy as np



# Create DataFrame
df_institutions = pd.DataFrame(institutions)

# Count affiliations per institution
counts = df_institutions['Institutions'].value_counts()

# Sort by count
counts = counts.sort_values(ascending=True)

# Plot
fig, ax = plt.subplots(figsize=(6.93, 6))  # 17.6 cm width = 6.93 in

# Generate colors
cmap = plt.get_cmap('tab20')
colors = cmap(np.linspace(0, 1, len(counts)))

bars = ax.barh(counts.index, counts.values, color=colors)

# Add count labels
for bar in bars:
    ax.text(bar.get_width() + 0.2, bar.get_y() + bar.get_height()/2,
            str(int(bar.get_width())), va='center', fontsize=9)

# Labels and title
ax.set_xlabel('Number of Affiliations per Institutions')
ax.set_title('Count of Affiliations by Institution')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.savefig(root_dir+'affiliations_chart.png', dpi=1200, bbox_inches='tight')
plt.show()
```
